# Fact Table for Sales

# View Schema of Tables being used

```python
df1 = spark.table("databricks_medallion_pipeline.silver.crm_sales")
print("Schema of CRM_Sales")
df1.printSchema()

df2 = spark.table("databricks_medallion_pipeline.gold.dim_products")
print("Schema of DIM_Products")
df2.printSchema()

df3 = spark.table("databricks_medallion_pipeline.gold.dim_customers")
print("Schema of DIM_Customers")
df3.printSchema()
```

# Transformation Logic

```python
query = """
SELECT
    sales.order_number,
    products.product_key,
    customers.customer_key,
    sales.order_date,
    sales.ship_date,
    sales.due_date,
    sales.sales_amount,
    sales.quantity,
    sales.price
FROM databricks_medallion_pipeline.silver.crm_sales sales
LEFT JOIN databricks_medallion_pipeline.gold.dim_products products
    ON sales.product_number = products.product_number
LEFT JOIN databricks_medallion_pipeline.gold.dim_customers customers
    ON sales.customer_id = customers.customer_id
"""

df = spark.sql(query)
```

```python
df.printSchema()
df.limit(10).display()
```

#Writing the Gold Table

```python
df.write.mode("overwrite").saveAsTable("databricks_medallion_pipeline.gold.fact_sales")
```
