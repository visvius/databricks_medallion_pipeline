#Ingestion Configuration

```python
INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/databricks_medallion_pipeline/bronze/source_system/source_crm/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/databricks_medallion_pipeline/bronze/source_system/source_crm/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/databricks_medallion_pipeline/bronze/source_system/source_crm/sales_details.csv",
        "table": "crm_sales_details"
    },
    {
        "source": "erp",
        "path": "/Volumes/databricks_medallion_pipeline/bronze/source_system/source_erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "erp",
        "path": "/Volumes/databricks_medallion_pipeline/bronze/source_system/source_erp/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "erp",
        "path": "/Volumes/databricks_medallion_pipeline/bronze/source_system/source_erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
    }
]
```

#Ingest files into Bronze Layer

```python
for item in INGESTION_CONFIG:
    source = item['source']
    path = item['path']
    table = item['table']

    print(f'Ingesting {source} -> Table: {table}')

    df = spark.read.option("header", "true")\
        .option("inferSchema", "true")\
        .csv(path)

    df.write\
        .format("delta")\
        .mode("overwrite")\
        .saveAsTable(f'databricks_medallion_pipeline.bronze.{table}')
```

#Delete ALL Tables

```python
catalog_name = "databricks_medallion_pipeline"
schema_name = "bronze"
tables_to_drop = [
    "crm_cust_info",
    "crm_prd_info",
    "crm_sales_details",
    "erp_cust_az12",
    "erp_loc_a101",
    "erp_px_cat_g1v2"
]

for table_name in tables_to_drop:
    full_table_path = f"{catalog_name}.{schema_name}.{table_name}"
    
    print(f"Dropping table: {full_table_path}")
    spark.sql(f"DROP TABLE IF EXISTS {full_table_path}")
```
