#Loading CRM CSV Files

```python
# cust_info.csv
df = spark.read.option("header", "true")\
        .option("inferSchema", "true")\
        .csv("/Volumes/databricks_medallion_pipeline/bronze/source_system/source_crm/cust_info.csv")
df.write.mode("overwrite").saveAsTable("databricks_medallion_pipeline.bronze.crm_cust_info")
df.display()
```
